Для определения области определения функции необходимо учесть следующие ограничения:

Извлечение квадратного корня может быть выполнено только для неотрицательного аргумента.\
Аргумент логарифма должен быть положительным.\
Исходя из этих ограничений, мы можем записать следующую систему неравенств:

$1-x^3 \geq 0 $(чтобы аргумент под корнем был неотрицательным)\
$y^2-1 > 0 $(чтобы аргумент логарифма был положительным)

Решим каждое из неравенств по отдельности:

$1-x^3 \geq 0$

Решаем уравнение 1-x^3 = 0:

$1 - x^3 = 0$\
$1 = x^3$\
$x = 1$

Функция не определена при x > 1, так как при этом значении корень из отрицательного числа не существует.\
Область определения функции для первого слагаемого будет (-∞, 1].

y^2 - 1 > 0

Решаем уравнение y^2 - 1 = 0:

y^2 = 1\
y = ±1

Функция не определена при y ≤ -1 или y ≥ 1, так как при этих значениях аргумент логарифма не положительный.\
Область определения функции для второго слагаемого будет (-∞, -1) ∪ (1, ∞).

Общая область определения функции будет пересечением этих двух областей, то есть (-∞, -1) ∪ (1, ∞) × (-1, 1).

Чтобы найти производную этой функции по $x$, сначала нам нужно применить правило дифференцирования сложной функции. Для этого мы разобьем функцию на внешнюю и внутреннюю функции:


$z=u^3$, где  $u=\left(1+\frac{\ln{x}}{\ln{y}}\right)$

Затем мы можем использовать правило дифференцирования степенной функции для нахождения производной $z$ по $u$:

$\frac{\partial z}{\partial u}=3u^2$

Затем мы можем применить правило дифференцирования сложной функции:


$\frac{\partial z}{\partial u}=\frac{\partial z}{\partial u}\times\frac{\partial u}{\partial x}$

Чтобы найти производную $\frac{du}{dx}$, мы можем использовать правило дифференцирования частного функций:

$\frac{\text{d}u}{\text{d}x}=\frac{\frac{1}{x}ln{y}}{\ln{y}}=\frac{1}{x\ln{y}}$

Теперь мы можем вычислить производную $z$ по $x$:

$\frac{\partial z}{\partial u}=\frac{\partial z}{\partial u}\cdot\frac{\partial u}{\partial x} = 3u^2\cdot\frac{1}{x\ln{y}}=\frac{3}{x\ln{y}}\cdot\left(1+\begin{array}{c}\ln{x}\\ \ln{y}\end{array}\right)^2$


Таким образом, производная $z$ по $x$ равна $\frac{3}{x \ln y} \left(1 + \frac{\ln x}{\ln y}\right)^2$.

Для нахождения полного дифференциала функции $z=f(x,y)=\sqrt{2xy+\cos\frac{x}{y}}$ в точке $(1,1)$, сначала найдём частные производные функции $f$ по $x$ и $y$:

$\frac{\text{d}f}{\text{d}x}=\frac{1}{2\sqrt{2xy+\cos\frac{x}{y}}}$


$\frac{\text{d}f}{\text{d}x}=\frac{1}{2\sqrt{2xy+\cos\frac{x}{y}}}\cdot\left(2y-\begin{array}{c}1\\ y^2\end{array}\cdot\sin\frac{x}{y}\right)$



$\frac{\text{d}f}{\text{d}y}=\frac{1}{2\sqrt{2xy+\cos\frac{x}{y}}}\cdot\left(2x+\begin{array}{c}1\\ y^2\end{array}\cdot\sin\frac{x}{y}\right)$

Затем подставим значения $x=1$ и $y=1$ в эти выражения:

$\frac{\text{d}f}{\text{d}x}=
|(1,1)=\frac{1}{2\sqrt{2\cdot1\cdot1+\cos1}}\cdot\left(2\cdot1-\frac{1}{1^2}\sin1\right)\approx0.4818$


$\frac{\text{d}f}{\text{d}y}=
|(1,1)=\frac{1}{2\sqrt{2\cdot1\cdot1+\cos1}}\cdot\left(2\cdot1+\frac{1}{1^2}\sin1\right)\approx0.7296$

Подставим найденные частные производные и приращения переменных $dx$ и $dy=1$ в формулу для полного дифференциала:

$dz=\frac{\text{d}f}{\text{d}x}|(1,1)$

$dx=\frac{\text{d}f}{\text{d}y}|(1,1)$

$dy\approx0.4818\cdot{dx}+0.7296\cdot{dy}$

Полный дифференциал функции $z=\sqrt{2xy+\cos\frac{x}{y}}$ в точке $(1,1)$ равен $dz \approx 0.4818\cdot dx + 0.7296\cdot dy$.

Для нахождения экстремума функции необходимо найти ее частные производные по каждой из переменных и приравнять их к нулю:

$\frac{\partial z}{\partial x} = 2x + y - 6 = 0$

$\frac{\partial z}{\partial y} = x + 2y - 9 = 0$

Решим данную систему уравнений методом подстановки. Из первого уравнения найдем $y$:

$y = 6 - 2x$

Подставим это выражение для $y$ во второе уравнение и решим его относительно $x$:

$x + 2(6-2x) - 9 = 0$

$x + 12 - 4x - 9 = 0$

$-3x + 3 = 0$

$x = 1$

Теперь найдем $y$:

$y = 6 - 2x = 6 - 2 \cdot 1 = 4$

Таким образом, найдены значения $x = 1$ и $y = 4$, которые являются стационарной точкой функции $z(x,y)$. Для того, чтобы проверить, является ли она точкой максимума или минимума, необходимо исследовать знаки вторых производных функции в этой точке.

Частные производные второго порядка:

$\frac{\partial^2 z}{\partial x^2} = 2$

$\frac{\partial^2 z}{\partial x \partial y} = 1$

$\frac{\partial^2 z}{\partial y^2} = 2$

Определитель матрицы равен:

$\begin{vmatrix}
2 & 1\\
1 & 2\\
\end{vmatrix} = 3 > 0$

Следовательно, стационарная точка $(1,4)$ является точкой минимума функции $z(x,y)$.

Ответ: $z_{min} = z(1,4) = 1^2 + 1\cdot4 + 4^2 - 6\cdot1 - 9\cdot4 = -29$.
