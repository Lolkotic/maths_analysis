# 1. Предложить пример функции, не имеющей предела в нуле и в бесконечностях.


sin(1/x)

В нуле: Функция sin(1/x) колеблется бесконечно часто в окрестности нуля, так как аргумент синуса стремится к бесконечности. Из-за этого колебания функция не приближается ни к одному конкретному значению, и поэтому предела в нуле не существует.

• В бесконечностях: Когда x стремится к плюс или минус бесконечности, аргумент синуса (1/x) стремится к нулю. Это значит, что sin(1/x) будет колебаться между -1 и 1, не приближаясь к определенному значению. Поэтому, предела в бесконечностях также не существует.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
np.seterr(divide='ignore')

x = np.arange(-100, 100, 0.5)
y = np.sin(1/x)

plt.axhline(linewidth=1.0, color="black")
plt.axvline(linewidth=1.0, color="black")

plt.plot(x,y)
plt.show()
```
